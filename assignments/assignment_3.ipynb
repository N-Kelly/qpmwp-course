Complete the function `turnover`.

```python

def turnover(self, return_series: pd.DataFrame, rescale: bool=True):

    dates = self.get_rebalancing_dates()
    to = {}
    to[dates[0]] = float(1)
    for count, rebalancing_date in enumerate(dates[1:]):

        previous_portfolio = self.get_previous_portfolio(rebalancing_date)
        current_portfolio = self.get_portfolio(rebalancing_date)
        
        if current_portfolio.rebalancing_date is None or previous_portfolio.rebalancing_date is None:
            raise ValueError('Portfolios must have a rebalancing date')

        if current_portfolio.rebalancing_date < previous_portfolio.rebalancing_date:
            raise ValueError('The previous portfolio must be older than the current portfolio')

        # Get the union of the ids of the weights in both portfolios (previous and current)
        previous_portfolio_weights = previous_portfolio.get_weights_series()
        current_portfolio_weights = current_portfolio.get_weights_series()

        previous_portfolio_weights_ids = previous_portfolio_weights.index
        current_portfolio_weights_ids = current_portfolio_weights.index

        union_weights_ids = previous_portfolio_weights_ids.union(current_portfolio_weights_ids)

        # Extend the weights of the previous portfolio to the the union of ids in both portfolios by adding zeros
        previous_portfolio_weights_aug = previous_portfolio_weights.reindex(union_weights_ids, fill_value=0)

        # Float the weights of the previous portfolio according to the price drifts in the market 
        # until the current rebalancing date
        # Hint: use function floating_weights from module portfolio.py
        floated_weights = floating_weights(return_series, previous_portfolio_weights_aug, start_date=dates[count], end_date=rebalancing_date, rescale=rescale)

        # Extract the weights of the current portfolio
        current_portfolio_weights_aug = current_portfolio_weights.reindex(union_weights_ids, fill_value=0)

        # Calculate the turnover
        to[rebalancing_date] = (current_portfolio_weights_aug - floated_weights.iloc[-1,:]).abs().sum()

    return pd.Series(to)
```

Complete the function `simulate`.

```python
def simulate(self,
                return_series: pd.DataFrame,
                fc: float = 0,
                vc: float = 0,
                n_days_per_year: int = 252) -> pd.Series:

    rebdates = self.get_rebalancing_dates()
    ret_list = []
    for rebdate in rebdates:
        next_rebdate = (
            rebdates[rebdates.index(rebdate) + 1]
            if rebdate < rebdates[-1]
            else return_series.index[-1]
        )

        portfolio = self.get_portfolio(rebdate)
        w_float = portfolio.float_weights(
            return_series=return_series,
            end_date=next_rebdate,
            rescale=False # Notice that rescale is hardcoded to False.
        )
        level = w_float.sum(axis=1)
        ret_tmp = level.pct_change(1)
        ret_list.append(ret_tmp)

    portf_ret = pd.concat(ret_list).dropna()

    if vc != 0:
        # Calculate turnover
        to = self.turnover(return_series=return_series,
                           rescale=False)
        # Calculate variable cost (vc) as a fraction of turnover and
        # subtract the variable cost from the returns at each rebalancing date
        VC = to*vc
        VC.index = pd.to_datetime(VC.index)
        VC_aug = VC.reindex(portf_ret.index, fill_value=0)
        portf_ret -= VC_aug

    if fc != 0:
        # Calculate number of days between returns
        # Calculate daily fixed cost based on the annual fixed cost (fc),
        # the number of days between two rebalancings and the number of days per year.
        # Subtract the daily fixed cost from the daily returns
        
        grouped_by_year = {year: ret for year, ret in portf_ret.groupby(portf_ret.index.year)}

        ret_list_fc = []

        for year, data in grouped_by_year.items():
            n_days = len(data.index)
            daily_fc = fc/n_days
            data -= daily_fc
            ret_list_fc.append(data)
        portf_ret = pd.concat(ret_list_fc)

    return portf_ret
```

3: Descriptive Statistics

(3 points)

Generate a table with descriptive statistics of the simulations. Include the following statistics:
- cumulative return (i.e., the return over the entire simulation period)
- annualized average returns
- annualized volatility
- sharpe ratio
- maximum drawdown

```python
from scipy.stats import gmean
# cumulative returns
cum_returns = (sim+1).cumprod()
final_cum_returns = cum_returns.iloc[-1,:] - 1

# grouping returns by year
grouped_by_year = {year: ret for year, ret in sim.groupby(sim.index.year)}

an_ret = pd.DataFrame(columns=sim.columns)
an_vol = pd.DataFrame(columns=sim.columns)

for year, returns in grouped_by_year.items():
    scalefactor = len(returns)

    # annualized average returns
    expected_return = ExpectedReturn(method='geometric', scalefactor=scalefactor)
    mu = expected_return.estimate(X=returns, inplace=False)
    an_ret.loc[len(an_ret)] = mu

    # annualized vol
    sigma = returns.std()*np.sqrt(scalefactor)
    an_vol.loc[len(an_vol)] = sigma

an_av_returns = pd.Series(gmean(1 + an_ret, axis=0) - 1, index = sim.columns)
an_av_vol = an_vol.mean(axis=0) # is arithmetic mean here alright?

# sharpe ratio
rf = 0.02 #  asumed yearly risk free rate
sharpe_ratio = (an_av_returns - rf)/an_av_vol

# maximum drawdwon
running_max = cum_returns.cummax(axis=0)
drawdowns = (cum_returns - running_max)/running_max

max_drawdown = drawdowns.min()

# putting all together
data = [final_cum_returns, an_av_returns, an_av_vol, sharpe_ratio, max_drawdown]
statistics = ['cummulative returns', 'annual average returns', 'annual average volatility', 'sharpe_ratio', 'max drawdown']
df = pd.DataFrame(data, columns=sim.columns, index=statistics)
df






```
